<b>문제[5] 타니타닉 데이터세트를 이용하여, 생존한 승객이 지금까지 살아 있을 때의 올 해 나이를 계산하여 새로운 'new_age' column으로 추가하고, 가장 나이가 많은 사람과 적은 사람의 나이차를 계산하여 출력하시오.  
 * 타이나틱 침몰 연도 : 1918년 (연도 차이에 의한 나이차이만 계산하세요.)
 * 생존여부 column : 'alive'  (yes: 생존, no: 사망)
 * 나이 column : 'age'

```python
df = sns.load_dataset('titanic')
```

```python
add_age = 2025-1918
print(f"added age : {add_age}")
df['new_age'] = np.where(df.alive=='yes', df.age+add_age, df.age)
df.dropna()
```

```python
print(f"oldest age : {df.new_age.max()}")
print(f"youngest age : {df.new_age.min()}")
print(f"oldest age - youngest age : {df.new_age.max()-df.new_age.min()}")
```

<b>문제[8] user_behavior_dataset.csv를 load하고 'Gender' 속성의 Male과 Female의 수를 bar로 나태니고, Male과 Female의 비율을 계산하여 pie 그래프로 나타내시오.  
 * bar와 pie 그래프를 하나의 화면(fig)에 출력하도록 작성하시오.
 * bar와 pie 그래프를 따로 출력할 경우 일부 감점.

```python
df = pd.read_csv('user_behavior_dataset.csv')
```

```python
gender_Value = df['Gender'].value_counts()

# 화면 분할 정의
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# 각 분할 영역에 그래프 작성하기
gender_Value.plot.bar(ax=axes[0])
gender_Value.plot.pie(ax=axes[1])

# 분할 그래프 화면에 나타내기 
plt.show()
```
