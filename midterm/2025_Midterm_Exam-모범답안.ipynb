<b>문제[3] 'cars.csv' 파일을 읽어(load) dataframe으로 변환하고,  "speed" 속성값이 10이상이면서 20미만인 데이터만 추출(필터링)하여 새로운 dataframe 객체에 할당하세요.
 * filtering된 dataframe을 출력하고 & 전체 수를 출력하세요.
 * dataframe 객체명을 자율적으로 정하세요.

```python
df = pd.read_csv('cars.csv')
filtered_df = df[(df.speed >= 10) & (df.speed < 20)]
filtered_df
```

```python
len(filtered_df)
```

<b>문제[4]<b> 위 문제 3번에서 추출(필터링)된 dataframe을 깊은복사 방식으로 copy하여 새로운 dataframe 객체를 생성하고, 'speed' 값이 짝수인 경우 해당 'dist' 속성값을 1 증가시키는 코드를 작성하세요.  
 * 복사 전/후 dataframe의 값의 변경 상태를 확인하세요.

```python
copy_df = filtered_df.copy()
add_value = copy_df.speed.apply(lambda x:1 if x % 2 == 0 else 0)
copy_df.dist = copy_df.dist + add_value
copy_df
```

<b>문제[8] user_behavior_dataset.csv를 load하고 'Gender' 속성의 Male과 Female의 수를 bar로 나태니고, Male과 Female의 비율을 계산하여 pie 그래프로 나타내시오.  
 * bar와 pie 그래프를 하나의 화면(fig)에 출력하도록 작성하시오.
 * bar와 pie 그래프를 따로 출력할 경우 일부 감점.

```python
df = pd.read_csv('user_behavior_dataset.csv')
gender_ratio = df.Gender.value_counts()
gender_ratio
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
gender_ratio.plot.bar(ax=axes[0])
gender_ratio.plot.pie(ax=axes[1], autopct='%1.0f%%')
plt.show()
```
