```python
# 01. 데이터셋에 대한 정보를 확인하시오. 몇 개의 행과 몇 개의 컬럼으로 구성되어 있는가?

print(f"rows : {len(df)}")
print(f"columns : {len(df.iloc[0])}")
```

```python
# 02. 데이터셋의 컬럼 이름을 출력하시오.

print(f"column names : {df.columns.to_list()}")
```

```python
# 03. titanic 데이터셋에서 앞쪽 10개의 행을 출력하시오.

print(f"titanic heads")
df.head(10)
```

```python
# 04. 선실은 몇 개의 등급으로 구성되어 있는지 확인하시오.

print(f"pclass types count : {df['pclass'].unique().size}") 
```

```python
# 05. 인덱스 기준 2행 4열의 값을 확인하시오.

print(f"row 2 column 4 : {df.iloc[1,3]}")
```

```python
# 06. 앞쪽 5명의 탑승객에 대해 선실, 성별, 나이, 생존 여부를 출력하시오.

df.head(5).loc[:, ['pclass','sex','age','alive']]
```

```python
# 07. 나이가 60세 이상인 승객의 성별, 나이, 생존 여부를 출력하시오.

df.loc[df['age']>=60,['sex','age','alive']]
```

```python
# 08. 나이가 10세 미만인 여성 승객의 나이, 생존 여부를 출력하시오.

df.loc[(df['age']<=10)&(df['sex']=='female'),['age','alive']]
```

```python
# 09. 생존한 남성의 평균 연령을 구하시오.

print(f"alive male mean age : {df.loc[(df['alive']=='yes')&(df['sex']=='male'),['age']].mean().values[0]}")
```

```python
# 10. 전체 승객 수 대비 생존자 비율을 구하시오.

print(f"alive rate : {len(df.loc[df['alive']=='yes'])/len(df)}")
```

```python
# 11. 생존한 승객 중 가장 나이가 어린 승객의 성별과 나이를 출력하시오.

df.loc[df['alive']=='yes'].sort_values(by='age').loc[:,['sex','age']].head(1)
```

```python
# 12. 1등실 승객 중 생존자 숫자를 확인하시오.

print(f"1class alive count : {len(df.loc[(df['alive']=='yes')&(df['pclass']==1)])}")
```

```python
# 13. 1등실 승객 수와 2등실 승객의 수 차이를 계산하시오.

print(f"1class - 2class : {len(df.loc[df['pclass']==1])-len(df.loc[df['pclass']==2])}")
```

```python
# 14. 생존한 3등실 승객 중 가장 나이가 많은 승객의 성별과 나이를 확인하시오.

df.loc[(df['alive']=='yes')&(df['pclass']==3)].sort_values(by='age',ascending=False).loc[:,['sex','age']].head(1)
```

```python
# 15. df에서 선실, 성별, 나이, 생존 여부 컬럼만 선택하여 df2에 복사한 후 내용을 출력하시오.

df2 = df.loc[:,['pclass','sex','age','alive']].copy()
df2
```

```python
# 16. 나이 등급(age_class) 컬럼을 df2에 추가하시오.
# 40세 미만이면 나이 등급은 ‘Low’, 40세 이상이면 ‘High’로 설정하고 추가된 내용을 확인하시오.

df2['age_class'] = np.where(df2['age'] < 40, 'Low', 'High')
df2.dropna()
```

```python
# 17. df2에서 탑승객이 모두 생존했다고 가정했을 때, 10년 후의 나이를 출력하시오.

pd.DataFrame(df2['age'] +10)
```

```python
# 18. df2에서 탑승객이 모두 생존했다고 가정했을 때, 10년 후의 평균 나이를 출력하시오.

print(f"10 years later mean alive age : {(df2['age']+10).mean()}")
```

```python
# 19. df2에서 선실 등급을 제외한 나머지 컬럼 데이터를 출력하시오.

df2.drop('pclass', axis=1)
```

```python
# 20. df2를 titanic.csv 파일에 저장하시오.

df2.to_csv('./titanic.csv')
```
