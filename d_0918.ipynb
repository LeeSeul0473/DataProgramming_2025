```python
# 데이터 초기화
sales = {"A": 781, "B": 650, "C": 705, "D": 406, "E": 580, "F": 450, "G": 550, "H": 640}
sales_index = ["A", "B", "C", "D", "E", "F", "G", "H"]

# Answer1
print("Answer 1")
for index in sales_index:
    print(sales[index])

# Answer2
print("Answer 2")
print(len(sales_index))

# Answer3
print("Answer 3")
print(sales[sales_index[1]])

# Answer4
print("Answer 4")
print(sales["F"])

# Answer5
print("Answer 5")
for i in range(2, 5):
    print(sales[sales_index[i]])

# Answer6
print("Answer 6")
print(sales["A"])
print(sales["B"])
print(sales["D"])

# Answer7
print("Answer 7")
for index in sales_index:
    if sales[index] < 500 or sales[index] > 700:
        print(f"{index}, {sales[index]}")

# Answer8
print("Answer 8")
sales_b = sales["B"]
for index in sales_index:
    if sales[index] > sales_b:
        print(f"{index}, {sales[index]}")

# Answer9
print("Answer 9")
for index in sales_index:
    if sales[index] < 600:
        print(index)

# Answer10
print("Answer 10")
for index in sales_index:
    if sales[index] < 600:
        sales_under_600 = sales[index]
        sales_under_600 = int(sales_under_600 + sales_under_600 * 0.2)
        print(f"{index} - Increase20% = {sales_under_600}")

# Answer11
print("Answer 11")
total_sum = sum(sales[index] for index in sales_index)
medium = total_sum // len(sales_index)
deviation = sum((medium - sales[index]) ** 2 for index in sales_index) // len(sales_index)
print(medium)
print(total_sum)
print(deviation)

# Answer12
print("Answer 12")
changed_sales = sales.copy()
print(sales)
changed_sales["A"]=810
changed_sales["C"]=820
print(changed_sales)

# Answer13
print("Answer 13")
sales["J"] = 400
print(sales)

# Answer14
print("Answer 14")
del sales["J"]
print(sales)

# Answer15
print("Answer 15")
sales2 = {index: sales[index] + 500 for index in sales_index}
print(sales)
print(sales2)
```
