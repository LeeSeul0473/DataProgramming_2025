# 평균, 중앙값, 절사평균
## 평균
```이름.mean()```
## 중앙값
* 데이터를 크기순으로 정렬했을 때, 가장 중앙에 위치하는 값
```이름.median()```
## 절사평균
* 데이터의 하위 n%와 상위 n%를 제외한 후, 남은 값들만 이용하여 평균을 계산하는 방식
```stats.trim_mean(이름, 0.2/*제외할퍼센트*/)```

# 사분위수(quantile)

* 주어진 데이터를 크기순으로 나열한 후, 이를 4등분하는 지점에 있는 값들
* 세 개의 값으로 전체 데이터의 특성을 추정하는 데 사용
* 4등분하기 위해 3개의 등분점이 필요
  * 제1사분위수(Q1)
  * 제2사분위수(Q2) : 중앙값과 동일
  * 제 3분위수(Q3)
* ```이름.quantile(/*퍼센트*/)```
  * ```이름.quantile([/*퍼센트 배열*/])```
  * ```이름.quantile([0.25,0.5,0.75])```

## 사분위수 포함 요약 정보 메서드
```이름.describe()```
* 출력 항목:
  * 값의 개수(count)
  * 평균(mean)
  * 표준편차(std)
  * 최소값(min)
  * 제1사분위수(Q1, 25%)
  * 중앙값(Q2, 50%)
  * 제3사분위수(Q3, 75%)
  * 최대값(max)

# 산포(distribution)
* 주어진 데이터의 값들이 퍼져 있는 정도(흩어져 있는 정도)
* 대표적인 지표: 분산(variance), 표준편차(standard deviation)
## 분산
```이름.var()```
## 표준편차
```이름.std()```
* 값이 작다면 → 학생들 간 성적(실력) 차이가 크지 않음을 의미한다.
* 값이 크다면 → 학생들 간 성적(실력) 차이가 크다는 의미로 해석할 수 있다.

# 히스토그램
* ```이름.plot.hist()```
* ```이름.plot.hist(bins=/*막대 개수*/,title='그래프 제목', xlabel='x축 레이블', ylabel='y축 레이블', color='막대 색')```
* 외관상 막대 그래프와 비슷한 형태의 그래프
  * 막대 그래프 : 범주형 데이터의 값 종류별 개수를 세어 작성
  * 히스토그램 : 연속형 데이터에서 특정 구간을 나누고, 해당 구간에 속하는 값들의 개수를 세는 방식
* 구간의 개수 기본값 : 10
* 구간 a~b에서 a은 이 구간에 포함이 안되고 b은 이 구간에 포함
* 히스토그램의 각 구간별 빈도수를 정확하게 알고 싶다면 value_counts() 메서드를 사용하면 된다.
  *  예시 : ```dist.value_counts(bins=6, sort=False)```
## 구간개수 관련
* 구간 크기가 너무 작으면 → 구간 개수가 많아지고, 각 구간별 빈도수가 작아져 유용한 정보를 전달하기 어려움
* 구간 개수가 너무 적으면 → 구간별 빈도수가 너무 커져 데이터의 변화를 파악하기 어려움

# 상자그림


* 상자 수염 그림(Box-and-Whisker Plot)이라고도 불린다.
* 사분위수를 시각적으로 표현한 그래프 형태이며, 데이터의 분포를 한눈에 파악할 수 있다.
* 단일변수 연속형 데이터의 분포를 이해하는 데 매우 유용하며, 하나의 그래프에서 다양한 정보를 전달할 수 있다.
* 최소값과 최대값은 정상 범위에 있는 관측값들 중에서 최소값과 최대값을 의미
  * 특이값(outlier)을 제외한 값들 중에서 최소값과 최대값
* 1사분위수(Q1)와 3사분위수(Q3) 사이가 상자로 표현
  * 전체 데이터의 50%가 위치하는 구간
* 모든 정보가 세로 방향으로 표현
  * 가로 방향은 별 의미가 없다.
  * 상자의 면적 역시 의미를 가지지 않는다.

````python
df = pd.read_csv('./data/iris.csv')
df.boxplot(column='Petal_Length', by='Species', grid=False)
plt.suptitle('')
plt.show()
````

* Setosa 품종의 꽃잎 길이가 가장 작고, Virginica 품종의 꽃잎 길이가 전반적으로 가장 크다.
* Setosa 품종은 값들이 좁은 범위에 집중되어 있어, 꽃잎 길이가 대부분 비슷함을 나타낸다.
* 반면, Versicolor와 Virginica 품종은 데이터의 분포가 넓게 퍼져 있어, 꽃잎 길이의 변동성이 크다는 것을 알 수 있다.

````python
favorite = pd.Series( ['WINTER', 'SUMMER', 'SPRING', 'SUMMER', 'SUMMER', 'FALL', 'FALL', 'SUMMER', 'SPRING', 'SPRING'])
fd = favorite.value_counts() # 도수분포를 fd 에 저장
fd.plot.bar(xlabel='Season',
ylabel='Frequency',
rot=30,
title='Favorite Season',
color='b',
grid=True, 
figsize=(6,4))
plt.subplots_adjust(bottom=0.2)
plt.show()
````

````python
fig, axes = plt.subplots(nrows=2, ncols=2)
df['Petal_Length'].plot.hist(ax=axes[0,0])
df['Petal_Length'].plot.box(ax=axes[0,1])
fd = df['Species'].value_counts() 
fd.plot.pie(ax=axes[1,0])
fd.plot.barh(ax=axes[1,1])
fig.suptitle('Multiple Graph Example', fontsize=14)
plt.show()
````
