```python
import pandas as pd
```

```python
df = pd.read_csv('../data/iris.csv')         # csv 파일 읽기
df                                           # 내용 확인
```

```python
# 데이터프레임 객체의 정보 확인
print(df.info())                                 # 배열의 정보
print('-'*50)
print(df.shape)                                  # 배열의 형태
print(df.shape[0])                               # 행의 개수
print(df.shape[1])                               # 컬럼의 개수
print('-'*50)
print(df.dtypes)                                 # 각 컬럼의 자료형
print('-'*50)
df['Species'] = df['Species'].astype('category')
print(df.dtypes)
print('-'*50)
print(df.columns)                                # 컬럼들의 이름 보기
print('-'*50)
print(df.head())                                 # 데이터 앞부분 일부 보기
print('-'*50)
print(df.tail())                                 # 데이터 뒷부분 일부 보기
print('-'*50)
print(df['Species'].unique())                    # 품종 정보 확인
```

```python
# 인덱싱과 슬라이싱
print(df.iloc[2,3])                              # 2행 3열의 값
print('-'*50)
print(df.loc[3,'Petal_Width'])                   # 'Petal_Width' 컬럼의 3행 값
print('-'*50)
print(df.loc[[0,2,4],['Petal_Length', 'Petal_Width']])
print('-'*50)
print(df.loc[5:8,'Petal_Length'])                # 'Petal_Length' 컬럼의 5~7행 값
print('-'*50)
print(df.iloc[:5,:4])                            # 0~4행, 0~3열의 값
print('-'*50)
print(df.loc[:, 'Petal_Length'])                 # 'Petal_Length' 컬럼의 모든 행의 값
print('-'*50)
print(df['Petal_Length'])                        # 'Petal_Length' 컬럼의 모든 값
print('-'*50)
print(df.Petal_Length)                           # 'Petal_Length' 컬럼의 모든 값
print('-'*50)
print(df.iloc[:5,:])                             # 0~4행의 모든 컬럼의 값
print('-'*50)
print(df.iloc[:5])                               # 0~4행의 모든 컬럼의 값
```

```python
print(df.loc[:, 'Petal_Length'])                  # 정식 표현
print('-'*50)
print(df['Petal_Length'])                         # 약식 표현
print('-'*50)
print(df.Petal_Length)                            # 약식 표현
print('-'*50)
print(df.iloc[:5,:])                              # 정식 표현
print('-'*50)
print(df.iloc[:5])                                # 약식 표현
```

```python
# 꽃잎의 길이가 6.5 이상인 행들의 모든 컬럼을 보이시오 
print(df.loc[df.Petal_Length >= 6.5, :])
print('-'*50)
print(df.loc[df.Petal_Length >= 6.5])            # 모든 컬럼의 경우 컬럼 인덱스 생략
print('-'*50)

# 꽃잎의 길이가 6.5 이상인 행들의 인덱스 번호를 보이시오 
print(df.loc[df.Petal_Length >= 6.5].index)
print('-'*50)

# 꽃잎의 길이가 3.5~3.8 사이인 행들의 모든 컬럼을 보이시오
print(df.loc[(df.Petal_Length >= 3.5) & 
       (df.Petal_Length <= 3.8)])
print('-'*50)

# 꽃잎의 길이가 1.3 미만이거나 6.5 를 초과하는 행들의 꽃잎의 길이와 폭을 보이시오
print(df.loc[(df.Petal_Length < 1.3) |
       (df.Petal_Length > 6.5), ['Petal_Length','Petal_Width']])
print('-'*50)

# where() 를 이용한 조건 검색
print(df.where(df.Petal_Length >= 6.5).dropna())
```

```python
# 데이터프레임 객체에 대한 산술 연산
print(df.loc[:, df.columns != 'Species'])               # 'Species' 컬럼을 제외
print('-'*50)
print(df.loc[:, ~df.columns.isin(['Species'])])         # 'Species' 컬럼을 제외
print('-'*50)
print(df.loc[:, df.columns != 'Species'] + 10)          # 모든 값들에 10을 더함 
print('-'*50)
print(df['Sepal_Length'] + df['Petal_Length'])          # 두 컬럼의 같은 행 값들끼리 연산
print('-'*50)

# 꽃잎의 길이가 5 이상인 경우는 꽃잎의 길이에 –1을 곱하시오 
# (5미만인 경우는 원래값 유지)
tmp = df['Petal_Length'].apply(lambda x: -1 if x >= 5 else 1)
print(df['Petal_Length'] * tmp) 
```

```python
df2 = df.loc[:, df.columns != 'Species']     # 숫자타입 컬럼만 추출
print(df2.sum())                                    # 값들의 합계
print('-'*50)
print(df2.mean())                                   # 값들의 평균
print('-'*50)
print(df2.median())                                 # 값들의 중앙값
print('-'*50)
print(df2.max())                                    # 값들의 최대값
print('-'*50)
print(df2.min())                                    # 값들의 최소값
print('-'*50)
print(df2.std())                                    # 값들의 표준편차
print('-'*50)
print(df2.var())                                    # 값들의 분산
print('-'*50)
print(df2.abs())                                    # 값들의 절대값
print('-'*50)
print(df2.describe())                               # 기초통계정보
```

```python
# 데이터프레임 객체에 있는 값의 수정
df3 = df.copy()                              # df를 df3으로 복사
df3.iloc[1,2] = 5.5                          # 1행2열의 값 수정
df3.loc[1,'Petal_Length'] = 1.1
print(df3.Petal_Length.to_list())                   # 시리즈를 리스트로
print('-'*50)
df3.loc[df.Petal_Length > 6.5,'Petal_Length'] *= 100
print(df3.Petal_Length.to_list())
```

```python
new_idx = df3.shape[0]                       # 추가할 행번호 결정
df3.loc[new_idx] = [1.1, 4.5, 3.4, 2.2, 'setosa']
print(df3.tail())
print('-'*50)

# 중간에 새로운 행 추가
new_row =  pd.DataFrame([[1.1, 2.2, 3.3, 4.4,'virginica']],
                        columns=df3.columns)
print(new_row)
print('-'*50)

df3 = pd.concat([df3.iloc[:10], new_row, df3.iloc[10:]], ignore_index = True)
print(df3.iloc[8:13,:])
print('-'*50)

# 여러 행의 추가
ext = pd.DataFrame([[1.2, 3.5, 4.3, 3.1, 'setosa'],
                    [2.1, 3.2, 2.3, 5.2, 'versicolor']],
                   columns=df3.columns)
print(ext)
print('-'*50)

df3 = df3._append(ext, ignore_index = True)
print(df3)
print('-'*50)

# 뒤쪽에 컬럼의 추가
new_col = df3.Petal_Length *10
df3['new_col'] = new_col
print(df3)
print('-'*50)

# 중간에 컬럼의 추가
df4 = df.copy()
df4.insert(loc=2, column='new_col2', value=new_col)            # 2번째 위치에 추가
print(df4)
```

```python
# 데이터프레임 객체에 대한 행과 컬럼의 삭제 
df5 = df.copy()

# 행의 삭제
df5 = df5.drop(index=[1,3])
print(df5)
print('-'*50)

df5 = df5.reset_index()
print(df5)
print('-'*50)

# 컬럼의 삭제
df5 = df5.drop(columns='Petal_Length')
print(df5)
print('-'*50)
```
