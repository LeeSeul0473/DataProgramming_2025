```python
import pandas as pd
import numpy as np
```

```python
# 자료의 입력
temp = pd.Series([-0.8, -0.1, 7.7, 13.8, 18.0, 22.4,
                  25.9, 25.3, 21.0, 14.0, 9.6, -1.4, np.nan])
print(temp)                                              # temp의 내용 확인
print('-'*50)
# 월이름을 레이블 인덱스로 지정하기
print(temp.index)                                       # 인덱스 내용 확인
print('-'*50)
temp.index = ['1월','2월','3월','4월',
              '5월','6월','7월','8월',
              '9월','10월','11월','12월', '평균']
print(temp)                                              # temp의 내용 확인
```

```python
print(temp.size)   # 배열의 크기(값의 개수)
print(len(temp))   # 배열의 크기(값의 개수)
print(temp.shape)  # 배열의 형태
print(temp.dtype)  # 배열의 자료형  
```

```python
# 인덱싱과 슬라이싱

print(temp.iloc[2])                                # 인덱스 2의 값 
print(temp.loc['3월'])                             # 인덱스 '3월'의 값 
print('-'*50)
print(temp.iloc[[3,5,7]])                          # 인덱스 3,5,7의 값 
print('-'*50)
print(temp.loc[['4월','6월','8월']])               # 인덱스 '4월','6월','8월'의 값
print('-'*50)
print(temp.iloc[5:8])                              # 인덱스 5~7의 값 
print('-'*50)
print(temp.loc['6월':'9월'])                       # 인덱스 '6월'~'9월'의 값 
print('-'*50)
print(temp.iloc[:4])                               # 인덱스 0~3의 값 
print('-'*50)
print(temp.iloc[9:])                               # 인덱스 9~11의 값
print('-'*50)
print(temp.iloc[:])                                # 인덱스 0~11의 값
```

```python
# 월평균 기온이 15도 이상인 월들의 기온을 보이시오
print(temp.loc[temp >= 15])
print('-'*50)

# 월평균 기온이 15도 이상 25도 미만인 월들의 기온을 보이시오
print(temp.loc[(temp >= 15) & (temp < 25)])
print('-'*50)

# 월평균 기온이 5도 미만이거나 25도 이상인 월들의 기온을 보이시오
print(temp.loc[(temp < 5) | (temp >= 25)])
print('-'*50)

# 3월보다 기온이 낮은 월들의 기온을 보이시오
march = temp.loc['3월']
print(temp.loc[temp < march])
print('-'*50)

# where() 를 이용한 조건 검색
print(temp.where(temp >= 15))
print('-'*50)
print(temp.where(temp >= 15).dropna())
```

```python
# 월평균 기온이 15도 이상인 월들의 이름을 보이시오

print(temp.loc[temp >= 15])            # 조건에 맞는 값 추출
print('-'*50)
print(temp.loc[temp >= 15].index)      # 조건에 맞는 값들의 인덱스 추출
```

```python
print(temp + 1)
print('-'*50)
print(2 * temp + 0.1)
print('-'*50)
print(temp + temp)
print('-'*50)
print(temp.loc[temp >= 15] + 1)     # 기온이 15도 넘는 월들에 대해서만 1 증가
```

```python
print(temp.sum())                                  # 값들의 합계
print(temp.mean())                                 # 값들의 평균
print(temp.median())                               # 값들의 중앙값
print(temp.max())                                  # 값들의 최대값
print(temp.min())                                  # 값들의 최소값
print(temp.std())                                  # 값들의 표준편차
print(temp.var())                                  # 값들의 분산
print('-'*50)
print(temp.abs())                                  # 값들의 절대값
print('-'*50)
print(temp.describe())                             # 기초통계정보
```

```python
salary = pd.Series([20, 15, 18, 30])        # 레이블 인덱스가 없는 시리즈 객체 
score = pd.Series([75, 80, 90, 60], 
                  index=['KOR', 'ENG', 'MATH', 'SOC'])
print(salary)
print('-'*50)
print(score)
print('-'*50)

# 값의 변경                               # 값 변경 시 절대 위치 index와 label index 모두 사용 가능
score.iloc[0] = 85                          # 인덱스 0 의 값을 변경 
print(score)
print('-'*50)

score.loc['SOC'] = 65                       # 인덱스 'SOC' 의 값을 변경 
print(score)
print('-'*50)
score.loc[['ENG','MATH']] = [70,80]         # 인덱스 'ENG','MATH' 의 값을 변경 
print(score)
```

```python
# 값의 추가 (레이블 인덱스가 있는 경우)
score.loc['PHY'] = 50                       # 없는 인덱스 추가
#score[＇PHY＇] = 50] #이 것도 가능
print(score)
#score.iloc[5] = 90                          # 에러 발생
print('-'*50)

# 값의 추가 (레이블 인덱스가 없는 경우)
next_idx = salary.size
#salary.iloc[next_idx] = 33                  # 에러 발생
salary.loc[next_idx] = 33                   # 정상 수행  #loc만으로 행 삽입 가능
print(salary)
print('-'*50)

# _append() 메서드를 이용한 값의 추가
new = pd.Series({'MUS': 95})
score._append(new)                          # score 변경 없음
print(score)
print('-'*50)
score = score._append(new)                  # score 변경됨
print(score)
print('-'*50)
salary._append(pd.Series([66]), ignore_index=True)
salary = salary._append(pd.Series([66]), ignore_index=True)
print(salary)
```

```python
# 값의 삭제
salary = pd.Series([20, 15, 18, 30])  # 레이블 인덱스가 없는 시리즈 객체
score = pd.Series([75, 80, 90, 60],
                  index=['KOR', 'ENG', 'MATH', 'SOC'])
print(score.drop('KOR'))                           # 레이블 인덱스가 있는 경우 
print('-'*50) 
print(score)
print('-'*50)                                    # score의 내용 변동 없음 
score = score.drop('ENG')
print(score)
print('-'*50)                                      # score의 내용 변경
salary = salary.drop(1)                     # 레이블 인덱스가 없는 경우
print(score)
```

```python
score_1 = pd.Series([80, 75, 90],
                    index=['KOR','ENG','MATH'])
score_2 = score_1            # score_1 과 score_2 는 동일한 객체
score_2.loc['KOR'] = 95
print(score_2)
print('-'*50) 
print(score_1)
print('-'*50) 
score_3 = score_1.copy()     # score_1 과 score_3 은 독립된 객체
score_3.loc['KOR'] = 70
print(score_3)
print('-'*50) 
print(score_1)
```
