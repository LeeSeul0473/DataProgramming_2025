```python
import pandas as pd
import numpy as np
```

```python
age = pd.Series([25, 34, 19, 45, 60]) # 직접 리스트 입력
print(age) # age의 내용 출력
print(type(age)) # age의 데이터형 확인 
```

```python
# 이미 생성된 리스트 객체를 입력
data = ['spring', 'summer', 'fall', 'winter']
season = pd.Series(data)
print(season)
print(season.iloc[2])  #인덱스 2의 값 
```

```python
score = pd.DataFrame([[85, 96, 40, 95],
                      [73, 69, 45, 80],
                      [78, 50, 60, 90]])
print(score)            # score 내용 출력 
print(type(score))      # score 데이터형 출력
print(score.index)      # 행 방향 인덱스 
print(score.columns)    # 열 방향 인덱스
print(score.iloc[1,2])  # 인덱스 1행 2열의 값
```

```python
# 넘파이 1차원 배열을 판다스로 변환

w_np = np.array([65.4, 71.3, np.nan, 57.8])   # 넘파이 1차원 배열
weight = pd.Series(w_np)                      # 넘파이 배열을 판다스 시리즈로
weight                                        # 판다스 시리즈
```

```python
# 판다스 시리즈를 넘파이로 변환

w_np2 = pd.Index.to_numpy(weight)             # 판다스 시리즈를 넘파이 배열로
w_np2  
```

```python
# 넘파이 2차원 배열로 부터 데이터프레임 생성

s_np = np.array([[85, 96, 40, 95],
                 [73, 69, 45, 80],
                 [78, 50, 60, 90]])
print(s_np)                                  # 넘파이 2차원 배열
score2 = pd.DataFrame(s_np)                  # 넘파이 배열을 데이터프레임으로
score2       
```

```python
# 데이터프레임을 넘파이 2차원 배열로 변환 

score_np = score2.to_numpy()                 # 데이터프레임을 넘파이 배열로
score_np     
```

```python
# 시리즈에 레이블 부여 

age = pd.Series([25, 34, 19, 45])
print(age)                                 # 레이블 부여 전
age.index = ['John','Jane','Tom','Luka']   # 행에 레이블 부여 
print(age)                                 # 레이블 부여 후
print(age.iloc[2])                         # 절대위치에 의한 인덱싱
print(age.loc['Tom'])                      # 레이블에 의한 인덱싱
```

```python
# 데이터프레임에 레이블 부여

score = pd.DataFrame([[85, 96, 40, 95],
                      [73, 69, 45, 80],
                      [78, 50, 60, 90]])
print(score)                                # 레이블 부여 전
score.index = ['John','Jane','Tom']         # 행에 레이블 부여 
score.columns = ['KOR','ENG','MATH','SCI']  # 열에 레이블 부여 
print(score)                                # 레이블 부여 후   
print(score.iloc[2,1])                      # 절대위치에 의한 인덱싱
print(score.loc['Tom','ENG'])               # 레이블에 의한 인덱싱
```

```python
age = pd.Series([25, 34, 19, 45, 60])
age.index = ['John','Jane','Tom','Micle','Tom']
print(age)
print(age.iloc[3])
print(age.loc['Tom'])
```
