```python
# 그룹이 있는 다중 산점도의 작성
df = pd.read_csv('./data/iris.csv')
dict = {'setosa':'red', 'versicolor':'green', 'virginica':'blue'}
colors = list(dict[key] for key in df.Species)  # 각 점의 색을 지정

df.plot.scatter(x='Petal_Length',  # x축 변수명
                y='Petal_Width',   # y축 변수명
                s=30,              # 점의 크기
                c=colors,          # 점의 색깔
                marker='o')        # 점의 모양
plt.show()
```

```python
fig, ax = plt.subplots()

for label, data in df.groupby('Species'):
    ax.scatter(x=data['Petal_Length'], y=data['Petal_Width'], s=30,
               c=dict[label], marker='o', label=label)
    print(data)

ax.set_xlabel('Petal_Length')
ax.set_ylabel('Petal_Length')
ax.legend()
fig.show()
```

```python
# 첫 번째 이미지
beers = [5,2,9,8,3,7,3,5,3,5]  # 음주량
bal = [0.1,0.03,0.19,0.12,0.04,0.0095,0.07,  # 혈중알콜 농도
       0.06,0.02,0.05]

dict = {'beers': beers, 'bal': bal}  # 딕셔너리 정의
df = pd.DataFrame(dict)  # 데이터프레임 생성

df.plot.scatter(x='beers', y='bal',  # 산점도
                title='Beers~Blood Alcohol Level')

# 두 번째 이미지
# 회귀선 계산
m, b = np.polyfit(beers, bal, 1)
# 회귀선 출력
plt.plot(beers, m * np.array(beers) + b)

plt.show()
```

```python
# 두 변수간 상관계수 계산
df['beers'].corr(df['bal'])  # 상관계수 계산

# 여러 변수들간 상관계수
df2 = pd.read_csv('./data/iris.csv')
df2.columns
df2 = df2.loc[:, ~df2.columns.isin(['Species'])]  # 품종 컬럼 제외
df2.columns

df2.corr()  # 상관계수 계산
```

```python
# 그래프에 한글 나타내기 준비
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

## (1) 데이터 준비 ---------------------------
df = pd.read_csv('./data/BostonHousing.csv')
df = df[['crim','rm','dis','tax','medv']]

titles = ['1인당 범죄율', '방의개수', '직업센터까지의 거리',
          '재산세','주택가격']

## (2) 그룹 컬럼 추가 ----------------------
grp = pd.Series(['M' for i in range(len(df))])
grp.loc[df.medv >= 25.0] = 'H'
grp.loc[df.medv <= 17.0] = 'L'
df['grp'] = grp

# 그룹 컬럼의 자료형과 레이블 순서 변경
new_odr = ['H','M','L']
new_dtype = CategoricalDtype(categories=new_odr, ordered=True)
df.grp = df.grp.astype(new_dtype)
df.grp.dtype
```

```python
# 화면 분할 정의
fig, axes = plt.subplots(nrows=2, ncols=3)
fig.subplots_adjust(hspace=0.5,wspace=0.3)  # 그래프 여백

# 각 분할 영역에 그래프 작성하기
for i in range(5):
    df[df.columns[i]].plot.hist(ax=axes[i//3,i%3],
                                ylabel="", xlabel=titles[i])

# 통합 그래프에 제목 지정
fig.suptitle('Histogram', fontsize=14)

# 분할 그래프 화면에 나타내기
fig.show()
```

```python

fig = plt.figure()
gs= gridspec.GridSpec(2,3)
fig.subplots_adjust(hspace=0.5,wspace=0.3)  # 그래프 여백

ax0_0 = fig.add_subplot(gs[0,0])
df[df.columns[0]].plot.hist(ax=ax0_0, ylabel='', xlabel=titles[0])

ax0_1 = fig.add_subplot(gs[0,1])
df[df.columns[1]].plot.hist(ax=ax0_1, ylabel='', xlabel=titles[1])

ax0_2 = fig.add_subplot(gs[0,2])
df[df.columns[2]].plot.hist(ax=ax0_2, ylabel='', xlabel=titles[2])

ax1_0 = fig.add_subplot(gs[1,:2])
df[df.columns[3]].plot.hist(ax=ax1_0, ylabel='', xlabel=titles[3])

ax1_1 = fig.add_subplot(gs[1,2])
df[df.columns[4]].plot.hist(ax=ax1_1, ylabel='', xlabel=titles[4])

fig.suptitle('Histogram', fontsize=14)

# 분할 그래프 화면에 나타내기
fig.show()
```

```python
## (5) 상자그림 ---------------------

fig, axes = plt.subplots(nrows=2, ncols=3)
fig.subplots_adjust(hspace=0.5,wspace=0.3)  # 그래프 여백

# 각 분할 영역에 그래프 작성하기
for i in range(5):
    df[df.columns[i]].plot.box(ax=axes[i//3,i%3],
                               label=titles[i])

fig.suptitle('Boxplot', fontsize=14)
fig.show()
```

```python
## (6) 그룹별 상자그림 -------------------

fig, axes = plt.subplots(nrows=2, ncols=3)
fig.subplots_adjust(hspace=0.5,wspace=0.3)  # 그래프 여백

# 각 분할 영역에 그래프 작성하기
for i in range(5):
    df.boxplot(column=df.columns[i], by='grp', grid=False,
               ax=axes[i//3,i%3], xlabel=titles[i])

fig.suptitle('Boxplot by group', fontsize=14)
fig.show()
```
